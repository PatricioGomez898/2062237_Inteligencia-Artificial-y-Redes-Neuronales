Parte 1: Carga y Limpieza

```python
#Importar librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Leer archivo CSV
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)
```

```python
#Visualizar datos y resumen estadístico
df.tail()
df.describe()
```

```python
#Eliminar filas donde dimensiones sean igual a 0
df = df[(df['x'] != 0) & (df['y'] != 0) & (df['z'] != 0)]
```

```python
#Cálculo de volumen espacial de cada pieza
df['volumen_aprox'] = df['x'] * df['y'] * df['z']
```

Parte 2: Visualización de datos

```python
#Gráfico de Caja (Boxplot)
#Distribución estadística de la columna price
plt.boxplot(df['price'])

plt.title("Distribución del Precio")
plt.xlabel("Precio")
plt.ylabel("Valores")

plt.show()
```

```python
#Gráfico de dispersión
#Grafica el volumen aproximado en el eje X contra el eje price en el eje Y
plt.scatter(df['volumen_aprox'], df['price'])

plt.title("Relación entre Volumen Aproximado y Precio")
plt.xlabel("Volumen Aproximado")
plt.ylabel("Precio")

plt.show()
```

```python
#Gráfico de Barras
#Muestra la cantidad total de piezas que existen para cada tipo de corte
cut_counts = df['cut'].value_counts()

plt.bar(cut_counts.index, cut_counts.values)

plt.title("Cantidad de Piezas por Tipo de Corte")
plt.xlabel("Tipo de Corte")
plt.ylabel("Cantidad de Piezas")

plt.show()
```
